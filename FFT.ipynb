```julia
using PyPlot
sampling = 2^21
nvec = 0: sampling-1
fsampling = 44100
t = nvec/fsampling;
freq1 = 100.0
freq2 = 300.0
freq3 = 500.0
freq = nvec*fsampling/sampling
dfreq = 1.0e0
wv = (sin(2*pi*freq1.*t) +sin(2*pi*freq2.*t)+sin(2*pi*freq3.*t))
for i=1:100
    freq1+=dfreq
    freq2+=dfreq
    freq3+=dfreq
    wv0 = (sin(2*pi*(freq1+rand(1)).*t)) +(sin(2*pi*(freq2+rand(1)).*t)) +(sin(2*pi*(freq3+rand(1)).*t)) 
    wv+=wv0
end
#wv1 = exp(2*pi*freq1*im.*t)+exp(-2*pi*freq1*im.*t)
#wv2 = exp(2*pi*freq2*im.*t)+exp(-2*pi*freq2*im.*t)
#wv = wv1+wv2
wave = wv #wv.^5+2*wv.^4+3*wv.^3-2*wv.^2+wv+1
#.*sin(2*pi*freq3*(sin(2*pi*freq1.*t)).*t); #100Hzと200Hzの信号
Y = fft(wave); #Nフーリエ変換により周波数領域に変換 
Pyy=10.0*log(abs(Y/sampling).^2); # Pyy－スペクトル密度conj－複素共役 
PyPlot.plt.figure()
PyPlot.plt.subplot(211);
#PyPlot.plt.xlim([0, 5])
PyPlot.plt.plot(t, wave)
PyPlot.plt.subplot(212);
PyPlot.plt.xlim([0, 700])
PyPlot.plt.ylim([-150, 10])
PyPlot.plt.plot(freq, Pyy)
```
